```python
# .txt to .csv & select 0-year-old female death rate

import os
import csv
from geopy.geocoders import Nominatim


folder_path = './data/death_rates'
output_folder = './data/csv'
os.makedirs(output_folder, exist_ok=True)
data_dict = {}
geolocator = Nominatim(user_agent="geoapiExercises")

for filename in os.listdir(folder_path):
    if filename.endswith('.txt'):
        file_path = os.path.join(folder_path, filename)
        with open(file_path, 'r') as file:
            lines = file.readlines()
            # country name
            country = lines[0].split(',')[0].strip()
            # latitude & longitude
            location = geolocator.geocode(country)
            if location:
                latitude = location.latitude
                longitude = location.longitude
            else:
                latitude = None
                longitude = None
            for line in lines[1:]:
                line_data = line.split()
                try:
                    # Year = 0, gender = female
                    year = int(line_data[0])
                    if int(line_data[1]) == 0:
                        female_baby_death_rate = float(line_data[2])
                        data_row = {
                            'Country': country,
                            'Year': year,
                            'Female Baby Death Rate': female_baby_death_rate,
                            'Latitude': latitude,
                            'Longitude': longitude
                        }
                        if year in data_dict:
                            data_dict[year].append(data_row)
                        else:
                            data_dict[year] = [data_row]
                except (ValueError, IndexError):
                    continue
sorted_years = sorted(data_dict.keys())
# create CSV
for year in sorted_years:
    data_rows = data_dict[year]
    output_file = os.path.join(output_folder, f'death_rates_{year}.csv')
    with open(output_file, 'w', newline='') as file:
        fieldnames = ['Country', 'Year', 'Female Baby Death Rate', 'Latitude', 'Longitude']
        writer = csv.DictWriter(file, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(data_rows)
#     with open(output_file, 'r') as file:
#         line_count = sum(1 for _ in file) - 1
#     print(f"finished: {output_file}，line count: {line_count}")
# # SELECT YEAR 1922-2022
```

```python
# combine all csv for 1 file

import os
import pandas as pd

folder_path = './data/csv'
merged_df = pd.DataFrame()
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(folder_path, filename)
        df = pd.read_csv(file_path)
        merged_df = merged_df.append(df, ignore_index=True)
output_file = './data/merged.csv'
merged_df.to_csv(output_file, index=False)

print(f"finished: {output_file}") 
```

```python
# .csv to .geojson & combine country border data from https://www.naturalearthdata.com/downloads/110m-cultural-vectors/
# some countries' name are not correspond. So ignore to save time

import geopandas as gpd
import pandas as pd


df = pd.read_csv('./data/merged.csv')
df = df[['Country', 'Year', 'Female Baby Death Rate']]
# Time datatype
df['Year'] = pd.to_datetime(df['Year'], format='%Y')
gdf = gpd.read_file('./data/countries/ne_110m_admin_0_countries.shp')
# "Country" & "NAME_EN"
df['Country'] = df['Country'].str.upper()
gdf['NAME_EN'] = gdf['NAME_EN'].str.upper()
merged_gdf = gdf.merge(df, left_on='NAME_EN', right_on='Country')
merged_gdf = merged_gdf[['Country', 'Year', 'Female Baby Death Rate', 'geometry']]
output_file = './data/filled_countries.geojson'
merged_gdf.to_file(output_file, driver='GeoJSON')

print(f"finished: {output_file}")
```
